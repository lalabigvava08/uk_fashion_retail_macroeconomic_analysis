Imports & paths

```python
import pandas as pd, numpy as np, re, os
import matplotlib.pyplot as plt

BRAND_FILES = {
    "Primark": "Data_Primark.xlsx",
    "Burberry": "Data_Burberry.xlsx",
    "Nobody's Child": "Data_NoCh.xlsx",
}
MACRO_PATH = "Macro_Data.xlsx"

os.makedirs("figs_tables", exist_ok=True)
```

Company extractor (robust to banner rows / weird headers)

```python
metric_row_regex = {
    "margin": r"profit\s*margin",
    "roe": r"(?:return\s*on\s*capital\s*employed|roce|roe)",
    "liquidity": r"liquidity\s*ratio",
    "gearing": r"gearing\s*\(%\)",
    "employees": r"(?:number\s*of\s*employees|employees)"
}

def _build_year_header(df, scan_rows=200):
    best_idx, best_cnt = None, -1
    scan = min(scan_rows, len(df))
    for i in range(scan):
        cnt = sum(bool(re.search(r'\b(19|20)\d{2}\b', str(v))) for v in df.iloc[i].values)
        if cnt > best_cnt:
            best_idx, best_cnt = i, cnt
    if best_idx is None or best_cnt <= 0:
        return None, None
    header, seen = [], {}
    for v in df.iloc[best_idx].values:
        s = "" if pd.isna(v) else str(v).strip()
        m = re.search(r'\b((?:19|20)\d{2})\b', s)
        lab = m.group(1) if m else s if s else None
        if lab is None: header.append(None); continue
        if lab in seen: seen[lab]+=1; header.append(f"{lab}_{seen[lab]}")
        else: seen[lab]=1; header.append(lab)
    body = df.drop(index=range(best_idx+1)).copy()
    body.columns = [f"col_{i}" if c is None else c for i, c in enumerate(header)]
    year_cols = [c for c in body.columns if re.fullmatch(r'(?:19|20)\d{2}(?:_\d+)?', str(c))]
    return body, year_cols

def extract_metrics_from_sheet(xlsx_path, sheet="Key financials & employees"):
    raw = pd.read_excel(xlsx_path, sheet_name=sheet, header=None)
    body, year_cols = _build_year_header(raw, scan_rows=200)
    if body is None or not year_cols: return None
    label_col = body.columns[0]
    labels = body[label_col].astype(str).str.lower().str.strip().fillna("")
    rows = {}
    for metric, rx in metric_row_regex.items():
        hit = body.loc[labels.str.contains(rx, regex=True)].head(1)
        if not hit.empty: rows[metric] = hit
    if not rows: return None
    out=[]
    for ycol in year_cols:
        m = re.search(r'\b((?:19|20)\d{2})\b', str(ycol))
        if not m: continue
        year = int(m.group(1)); rec={"year": year}
        for metric, r in rows.items():
            val = r[ycol].iloc[0] if ycol in r.columns else np.nan
            try: val = float(str(val).replace(',','').replace('%',''))
            except: val = np.nan
            rec[metric]=val
        out.append(rec)
    return pd.DataFrame(out).sort_values("year").reset_index(drop=True)

brand_metrics = {}
for brand, path in BRAND_FILES.items():
    df = extract_metrics_from_sheet(path, "Key financials & employees")
    if df is None:
        df = extract_metrics_from_sheet(path, "Key information")
    brand_metrics[brand] = df
```

Macro extractors (GDP, Unemp, Bank Rate, CPIH clothing, Retail TCF)

```python
# GDP (ONS chained volume measures, £m)
def extract_gdp_annual(xlsx_path, sheet='GDP '):
    df = pd.read_excel(xlsx_path, sheet_name=sheet, header=None)
    yr = df.iloc[:,0].astype(str).str.fullmatch(r'(19|20)\d{2}')
    year_rows = df[yr].copy()
    num_cols = [c for c in year_rows.columns if pd.to_numeric(year_rows[c], errors='coerce').notna().mean()>0.5]
    gdp_col = num_cols[-1]
    out = year_rows[[0, gdp_col]]
    out.columns=['year','gdp_millions_cvm']
    out['year']=out['year'].astype(int)
    out['gdp_millions_cvm']=pd.to_numeric(out['gdp_millions_cvm'], errors='coerce')
    return out.dropna().sort_values('year')

# Unemployment (3-month rolling) -> calendar-year average
def extract_unemployment_annual(xlsx_path, sheet='Unemployment rate '):
    df = pd.read_excel(xlsx_path, sheet_name=sheet, header=0)
    df = df.rename(columns={df.columns[0]:'period', df.columns[1]:'unemployment_rate'})
    def parse_period(p):
        m = re.search(r'([A-Za-z]{3})[-\s]?([A-Za-z]{3})\s+(\d{4})', str(p))
        if not m: return None
        end_mon, year = m.group(2).title(), int(m.group(3))
        dt = pd.to_datetime(f"01 {end_mon} {year}") + pd.offsets.MonthEnd(0)
        return dt
    df['date']=df['period'].apply(parse_period)
    df = df.dropna(subset=['date','unemployment_rate'])
    df['year']=df['date'].dt.year
    return df.groupby('year', as_index=False)['unemployment_rate'].mean()

# Bank Rate (list of changes) -> end-of-year rate
def extract_bankrate_eoy(xlsx_path, sheet='Bank Rate data'):
    raw = pd.read_excel(xlsx_path, sheet_name=sheet, header=None)
    rows=[]
    for i in range(len(raw)):
        vals=[v for v in raw.iloc[i].tolist() if pd.notna(v)]
        if len(vals)>=2:
            rows.append((vals[-2], vals[-1]))
    b = pd.DataFrame(rows, columns=['date_str','rate'])
    b['date']=pd.to_datetime(b['date_str'], dayfirst=True, errors='coerce')
    b['rate']=pd.to_numeric(b['rate'], errors='coerce')
    b=b.dropna().sort_values('date'); b['year']=b['date'].dt.year
    eoy = b.groupby('year', as_index=False).tail(1)[['year','rate']].drop_duplicates('year')
    return eoy.sort_values('year').rename(columns={'rate':'bank_rate_eoy'})

# CPIH clothing & footwear monthly index -> calendar-year average
def extract_cpih_clothing_annual(xlsx_path, sheet='CPIH'):
    df = pd.read_excel(xlsx_path, sheet_name=sheet, header=None)
    name_row_idx=None
    for i in range(len(df)):
        if any((str(v).strip().lower()=='name') for v in df.iloc[i].tolist()):
            name_row_idx=i; break
    name_row=[str(v).strip() if pd.notna(v) else "" for v in df.iloc[name_row_idx].tolist()]
    try:
        clothing_col = next(idx for idx,v in enumerate(name_row) if v.upper().startswith('CLOTHING AND FOOTWEAR'))
    except StopIteration:
        clothing_col=2
    idxrow=[str(v).strip() if pd.notna(v) else "" for v in df.iloc[name_row_idx-1].tolist()]
    try:
        indexdate_col = next(idx for idx,v in enumerate(idxrow) if v.lower()=='index date')
    except StopIteration:
        indexdate_col=0
    data=df.iloc[name_row_idx+1:].copy()
    data=data[data.iloc[:,indexdate_col].astype(str).str.fullmatch(r'\d{6}')]
    data['yyyymm']=data.iloc[:,indexdate_col].astype(str); data['year']=data['yyyymm'].str[:4].astype(int)
    data['clothing_index']=pd.to_numeric(data.iloc[:, clothing_col], errors='coerce')
    return data.dropna(subset=['clothing_index']).groupby('year', as_index=False)['clothing_index'].mean()\
               .rename(columns={'clothing_index':'cpih_clothing_index_avg'})

# Retail Sales (Textile, Clothing & Footwear) -> calendar-year average
def extract_retail_tcf_annual(xlsx_path, sheet=' Average Weekly Retail Sales '):
    df = pd.read_excel(xlsx_path, sheet_name=sheet, header=None)
    out=[]
    for i in range(len(df)):
        row=df.iloc[i].tolist()
        for j,v in enumerate(row):
            s=str(v).strip()
            if re.fullmatch(r'\d{4}[A-Za-z]{3}', s):
                # next non-null to the right is the value
                val=None
                for k in range(j+1, len(row)):
                    if pd.notna(row[k]):
                        try: val=float(str(row[k]).replace(',','')); break
                        except: pass
                if val is not None: out.append((int(s[:4]), val))
                break
    return pd.DataFrame(out, columns=['year','retail_tcf_avg_weekly_thou']).groupby('year', as_index=False).mean()

# run macro loaders
gdp  = extract_gdp_annual(MACRO_PATH)
unem = extract_unemployment_annual(MACRO_PATH)
bank = extract_bankrate_eoy(MACRO_PATH)
cpih = extract_cpih_clothing_annual(MACRO_PATH)
tcf  = extract_retail_tcf_annual(MACRO_PATH)

macro = gdp.merge(unem, on='year', how='outer')\
           .merge(bank, on='year', how='outer')\
           .merge(cpih, on='year', how='outer')\
           .merge(tcf, on='year', how='outer')
```

Build the Performance Index per brand & merge with macro

```python
def zscore(x):
    x = x.astype(float)
    s = np.nanstd(x)
    return (x - np.nanmean(x)) / (s if s not in (0, np.nan) else 1.0)

panel_list=[]
for brand, df in brand_metrics.items():
    d=df.copy(); d['brand']=brand
    for col in ['margin','roe','liquidity','gearing']:
        d[f'z_{col}']=zscore(d[col]) if col in d else np.nan
    d['z_gearing_neg'] = -d['z_gearing']   # higher gearing = worse
    d['perf_index'] = d[['z_margin','z_roe','z_liquidity','z_gearing_neg']].mean(axis=1, skipna=True)
    panel_list.append(d[['year','brand','margin','roe','liquidity','gearing','employees','perf_index']])

brand_panel = pd.concat(panel_list, ignore_index=True)
panel = brand_panel.merge(macro, on='year', how='left').sort_values(['brand','year']).reset_index(drop=True)
panel.head(10)
```

Correlations & figures (relationship-first)

```python
# Correlation with macro, per brand
macro_cols = ['gdp_millions_cvm','unemployment_rate','bank_rate_eoy','cpih_clothing_index_avg','retail_tcf_avg_weekly_thou']
corr_frames=[]
for brand in panel['brand'].unique():
    sub = panel[panel['brand']==brand][['perf_index']+macro_cols].dropna()
    if sub.empty: continue
    corr = sub.corr(numeric_only=True).loc[['perf_index'], macro_cols].T.reset_index()
    corr.columns=['macro_variable','corr_with_perf_index']; corr['brand']=brand
    corr_frames.append(corr)
corr_table = pd.concat(corr_frames, ignore_index=True)
corr_table.to_csv("figs_tables/correlations_perf_vs_macro.csv", index=False)

# Overlays (z-scored) and simple scatter + linear fit
def _z(v): return (v - np.nanmean(v)) / (np.nanstd(v) if np.nanstd(v) else 1.0)

def plot_overlay(sub, macro_col, macro_label, out_png):
    df=sub[['year','perf_index',macro_col]].dropna().copy()
    if df.empty: return
    df['z_perf']=_z(df['perf_index']); df['z_macro']=_z(df[macro_col])
    plt.figure(figsize=(8,5))
    plt.plot(df['year'], df['z_perf'], marker='o', label='Performance index (z)')
    plt.plot(df['year'], df['z_macro'], marker='s', label=f'{macro_label} (z)')
    plt.title(f"{sub['brand'].iloc[0]}: Performance vs {macro_label} (z-scored)")
    plt.xlabel("Year"); plt.ylabel("Standardised units"); plt.legend(); plt.tight_layout()
    plt.savefig(out_png, dpi=160); plt.close()

def plot_scatter(sub, macro_col, macro_label, out_png):
    df=sub[['perf_index', macro_col]].dropna().copy()
    if df.empty: return
    x=df[macro_col].values; y=df['perf_index'].values
    coef=np.polyfit(x,y,1); fx=np.poly1d(coef)
    xs=np.linspace(x.min(), x.max(), 100)
    plt.figure(figsize=(6,5))
    plt.scatter(x,y); plt.plot(xs, fx(xs))
    plt.title(f"{sub['brand'].iloc[0]}: Performance vs {macro_label}")
    plt.xlabel(macro_label); plt.ylabel("Performance index"); plt.tight_layout()
    plt.savefig(out_png, dpi=160); plt.close()

labels = {
    'gdp_millions_cvm': 'GDP (CVM, £m)',
    'unemployment_rate': 'Unemployment rate (%)',
    'bank_rate_eoy': 'Bank Rate (end-year)',
    'cpih_clothing_index_avg': 'CPIH Clothing & Footwear index',
    'retail_tcf_avg_weekly_thou': 'Retail Sales TCF (avg weekly £k)'
}
for brand in panel['brand'].unique():
    sub = panel[panel['brand']==brand].copy()
    # overlays
    for col, lab in labels.items():
        plot_overlay(sub, col, lab, f"figs_tables/{brand}_perf_vs_{col}.png")
    # scatter for two headline series
    plot_scatter(sub, 'gdp_millions_cvm', 'GDP (CVM, £m)', f"figs_tables/{brand}_scatter_perf_gdp.png")
    plot_scatter(sub, 'unemployment_rate', 'Unemployment rate (%)', f"figs_tables/{brand}_scatter_perf_unemp.png")

# small preview
panel.head(50).to_csv("figs_tables/panel_preview.csv", index=False)
"Done – see figs_tables/ for PNGs and CSVs."
```
